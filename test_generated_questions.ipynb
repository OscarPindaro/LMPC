```python
def format_sample(sample, question_word_limit=20, answer_word_limit=20):
    question = '\n'.join([' '.join(sample.question.split()[i:i+question_word_limit]) for i in range(0, len(sample.question.split()), question_word_limit)])
    answer = '\n'.join([' '.join(sample.answer.split()[i:i+answer_word_limit]) for i in range(0, len(sample.answer.split()), answer_word_limit)])
    character = sample.meta["character"]
    return f"{character}:\n{question}\n\nOrlando Marlo: {answer}"
```
